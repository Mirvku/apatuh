# Import Libraries & Load Iamge

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import urllib.request
```

# Download & Load Image

```python
img = cv2.imread("IKANN.jpeg")

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

```python
gray = cv2.imread('IKANN.jpeg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.figure(figsize=(6,5))
plt.imshow(gray, cmap='gray')
plt.title('Gambar Asli')
plt.axis('off')

plt.show()
```

# Sobel

```python
s_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
s_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)
s_mag = cv2.magnitude(s_x, s_y)

sobel_x = cv2.convertScaleAbs(s_x)
sobel_y = cv2.convertScaleAbs(s_y)
sobel_magnitude = cv2.convertScaleAbs(s_mag)
```

```python
plt.figure(figsize=(15,4))

plt.subplot(1,3,1)
plt.imshow(sobel_x, cmap='gray')
plt.title('Sobel X')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(sobel_y, cmap='gray')
plt.title('Sobel Y')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(sobel_magnitude, cmap='gray')
plt.title('Sobel Magnitude')
plt.axis('off')

plt.show()
```

# Canny

```python
canny = cv2.Canny(gray, 75, 200)
```

```python
plt.figure(figsize=(6,5))
plt.imshow(canny, cmap='gray')
plt.title('Canny')
plt.axis('off')

plt.show()
```

# Orde Kedua

```python
blur = cv2.GaussianBlur(gray, (5,5), sigmaX=0)
lap = cv2.Laplacian(blur,cv2.CV_64F)
log = cv2.convertScaleAbs(lap)
```

```python
plt.figure(figsize=(6,5))
plt.imshow(log, cmap='gray')
plt.title('Log (Laplacian + Gaussian)')
plt.axis('off')

plt.show()
```
